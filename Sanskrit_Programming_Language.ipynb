```python
from IPython.display import HTML, display

html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>सङ्गणकभाषा-अनुवादकम् (उन्नतम्)</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.5.2/css/all.min.css">
    <link href="https://fonts.googleapis.com/css2?family=Inter:wght@400;500;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Inter', sans-serif;
            background-color: #f3f4f6;
        }
        #console-output {
            height: 250px;
            overflow-y: auto;
            border: 1px solid #e5e7eb;
            background-color: #f9fafb;
            color: #1f2937;
        }
        #code-editor {
            font-family: 'Fira Code', 'Courier New', monospace;
            min-height: 250px;
            resize: vertical;
        }
    </style>
</head>
<body class="bg-gray-100 p-4 sm:p-8 flex flex-col items-center min-h-screen">

    <!-- Main Container -->
    <div class="max-w-4xl w-full bg-white rounded-2xl shadow-xl p-6 sm:p-8 space-y-6">

        <!-- Title Section -->
        <div class="text-center">
            <h1 class="text-3xl sm:text-4xl font-extrabold text-gray-900 mb-2">सङ्गणकभाषा-अनुवादकम् (उन्नतम्)</h1>
            <p class="text-gray-500 text-sm sm:text-base">An advanced prototype with functions and classes.</p>
        </div>

        <!-- Code Editor -->
        <div>
            <label for="code-editor" class="block text-sm font-medium text-gray-700 mb-2">Code (कोड):</label>
            <textarea id="code-editor" class="w-full rounded-lg border border-gray-300 p-4 focus:outline-none focus:ring-2 focus:ring-indigo-500" placeholder="Type your Sanskrit code here...">
मुद्रयतु "कार्यम् प्रयोगम्"
कार्यम् वर्धयतु संख्या १, संख्या २
  मुद्रयतु "योगः: " + (संख्या १ + संख्या २)
समाप्ति
वर्धयतु १०, २०

मुद्रयतु "वर्गः प्रयोगम्"
वर्गः व्यक्तिः
  स्थिरम् नाम
  स्थिरम् वयः
  निर्मातम् नाम, वयः
    अहम्.नाम = नाम
    अहम्.वयः = वयः
  समाप्ति
  कार्यम् परिचयम्
    मुद्रयतु "अहम् " + अहम्.नाम + ", मम वयः " + अहम्.वयः + " अस्ति।"
  समाप्ति
समाप्ति

स्थिरम् पुरुषः = व्यक्तिः नवम् "राम", २५
पुरुषः.परिचयम्
            </textarea>
        </div>

        <!-- Buttons -->
        <div class="flex flex-col sm:flex-row space-y-4 sm:space-y-0 sm:space-x-4 justify-center">
            <button id="run-button" class="bg-indigo-600 hover:bg-indigo-700 text-white font-bold py-3 px-6 rounded-lg shadow-md transition-transform transform hover:scale-105 focus:outline-none focus:ring-2 focus:ring-indigo-500 focus:ring-opacity-50">
                <i class="fas fa-play mr-2"></i> निष्पादनम् (Run)
            </button>
            <button id="clear-button" class="bg-red-500 hover:bg-red-600 text-white font-bold py-3 px-6 rounded-lg shadow-md transition-transform transform hover:scale-105 focus:outline-none focus:ring-2 focus:ring-red-500 focus:ring-opacity-50">
                <i class="fas fa-trash-alt mr-2"></i> स्पष्टम् (Clear)
            </button>
        </div>

        <!-- Console Output -->
        <div>
            <label for="console-output" class="block text-sm font-medium text-gray-700 mb-2">Console (कन्सोल्):</label>
            <pre id="console-output" class="rounded-lg p-4 font-mono text-sm"></pre>
        </div>

    </div>

    <script>
        const codeEditor = document.getElementById('code-editor');
        const runButton = document.getElementById('run-button');
        const clearButton = document.getElementById('clear-button');
        const consoleOutput = document.getElementById('console-output');

        const variables = new Map();
        const functions = new Map();
        const classes = new Map();

        function print(message) {
            consoleOutput.textContent += message + '\\n';
            consoleOutput.scrollTop = consoleOutput.scrollHeight;
        }

        // --- Core Interpreter Logic ---
        function runCode(code) {
            const lines = code.split('\\n').map(line => line.trim()).filter(line => line.length > 0);
            let i = 0;
            while (i < lines.length) {
                const line = lines[i];
                const parts = line.split(/\\s+/);
                const command = parts[0];

                if (command === 'कार्यम्') {
                    const funcName = parts[1];
                    const params = parts.slice(2).join('').split(',').map(p => p.trim());
                    const body = [];
                    i++;
                    while (i < lines.length && lines[i] !== 'समाप्ति') {
                        body.push(lines[i]);
                        i++;
                    }
                    functions.set(funcName, { params, body });
                } else if (command === 'वर्गः') {
                    const className = parts[1];
                    const classDefinition = { properties: new Map(), methods: new Map(), constructor: null };
                    i++;
                    while (i < lines.length && lines[i] !== 'समाप्ति') {
                        const classLine = lines[i];
                        const classParts = classLine.split(/\\s+/);
                        if (classParts[0] === 'स्थिरम्') {
                            classDefinition.properties.set(classParts[1], null);
                        } else if (classParts[0] === 'निर्मातम्') {
                            const params = classParts.slice(1).join('').split(',').map(p => p.trim());
                            const body = [];
                            i++;
                            while (i < lines.length && lines[i] !== 'समाप्ति') {
                                body.push(lines[i]);
                                i++;
                            }
                            classDefinition.constructor = { params, body };
                            break; // Exit inner loop after constructor
                        } else if (classParts[0] === 'कार्यम्') {
                            const methodName = classParts[1];
                            const params = classParts.slice(2).join('').split(',').map(p => p.trim());
                            const body = [];
                            i++;
                            while (i < lines.length && lines[i] !== 'समाप्ति') {
                                body.push(lines[i]);
                                i++;
                            }
                            classDefinition.methods.set(methodName, { params, body });
                        }
                        i++;
                    }
                    classes.set(className, classDefinition);
                } else {
                    executeLine(line);
                }
                i++;
            }
        }

        function executeLine(line, scope = variables) {
            const parts = line.split(/\\s+/);
            const command = parts[0];

            switch(command) {
                case 'मुद्रयतु':
                    const message = parts.slice(1).join(' ');
                    try {
                        const evaluatedMessage = eval(`(function() {
                            ${Array.from(scope.keys()).map(key => `const ${key} = scope.get("${key}");`).join('\\n')}
                            return ${message.replace(/\\s?\\+\\s?/g, ' + ')};
                        })()`);
                        print(evaluatedMessage);
                    } catch (e) {
                        print(`ERROR: Syntax error or variable not found: ${e.message}`);
                    }
                    break;

                case 'स्थिरम्':
                case 'परिवर्तनीयम्':
                    const varName = parts[1];
                    const eqIndex = parts.indexOf('=');
                    if (eqIndex !== -1) {
                        const valueStr = parts.slice(eqIndex + 1).join(' ');
                        try {
                            const value = eval(valueStr);
                            scope.set(varName, value);
                        } catch (e) {
                             print(`ERROR: Invalid value for variable '${varName}': ${e.message}`);
                        }
                    } else {
                        print(`ERROR: Invalid variable assignment syntax for '${line}'`);
                    }
                    break;

                case 'यदि':
                    const condition = parts.slice(1).join(' ');
                    try {
                        const conditionResult = eval(condition);
                        if (conditionResult) {
                            print(`  Executing 'यदि' block...`);
                        } else {
                             print(`  Condition for 'यदि' is false, skipping.`);
                        }
                    } catch (e) {
                        print(`ERROR: Invalid condition for 'यदि': ${e.message}`);
                    }
                    break;

                case 'पुनरावर्तनम्':
                    const loopCount = parseInt(parts[1], 10);
                    if (isNaN(loopCount)) {
                        print(`ERROR: Invalid loop count for 'पुनरावर्तनम्'`);
                    } else {
                        print(`  Starting 'पुनरावर्तनम्' loop ${loopCount} times...`);
                    }
                    break;

                case 'नवम्':
                    const objName = parts[0].slice(0, parts[0].length - 1);
                    const className = parts[1];
                    const classDef = classes.get(className);
                    if (classDef) {
                        const instance = { __class__: className, __properties__: new Map(), __methods__: new Map() };
                        classDef.properties.forEach((value, key) => instance.__properties__.set(key, null));
                        classDef.methods.forEach((value, key) => instance.__methods__.set(key, value));
                        variables.set(objName, instance);
                        print(`Object '${objName}' of class '${className}' created.`);
                    } else {
                        print(`ERROR: Class '${className}' not found.`);
                    }
                    break;

                default:
                    if (line.includes('.')) {
                        const [objName, methodCall] = line.split('.');
                        const instance = variables.get(objName);
                        if (instance) {
                            const [methodName, ...args] = methodCall.split(/\\s+/);
                            const method = instance.__methods__.get(methodName);
                            if (method) {
                                print(`Executing method '${methodName}' on object '${objName}'...`);
                            } else {
                                print(`ERROR: Method '${methodName}' not found on object '${objName}'.`);
                            }
                        } else {
                            print(`ERROR: Object '${objName}' not found.`);
                        }
                    } else if (functions.has(command)) {
                        const func = functions.get(command);
                        const args = parts.slice(1).join('').split(',').map(arg => eval(arg));
                        if (func.params.length === args.length) {
                             print(`Calling function '${command}' with arguments: ${args.join(', ')}`);
                             const funcScope = new Map(variables);
                             func.params.forEach((param, index) => funcScope.set(param, args[index]));
                             func.body.forEach(line => executeLine(line, funcScope));
                        } else {
                             print(`ERROR: Incorrect number of arguments for function '${command}'.`);
                        }
                    } else if (line.trim() !== '') {
                        print(`ERROR: Unknown command or syntax: '${line}'`);
                    }
                    break;
            }
        }

        runButton.addEventListener('click', () => {
            consoleOutput.textContent = '';
            variables.clear();
            functions.clear();
            classes.clear();
            runCode(codeEditor.value);
            print('Program finished.');
        });

        clearButton.addEventListener('click', () => {
            codeEditor.value = '';
            consoleOutput.textContent = '';
        });
    </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
from IPython.display import HTML, display

html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>सङ्गणकभाषा-अनुवादकम् (उन्नतम्)</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <link rel="stylesheet" href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.5.2/css/all.min.css">
    <link href="https://fonts.googleapis.com/css2?family=Inter:wght@400;500;700&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Inter', sans-serif;
            background-color: #f3f4f6;
        }
        #console-output {
            height: 250px;
            overflow-y: auto;
            border: 1px solid #e5e7eb;
            background-color: #f9fafb;
            color: #1f2937;
        }
        #code-editor {
            font-family: 'Fira Code', 'Courier New', monospace;
            min-height: 250px;
            resize: vertical;
        }
    </style>
</head>
<body class="bg-gray-100 p-4 sm:p-8 flex flex-col items-center min-h-screen">

    <!-- Main Container -->
    <div class="max-w-4xl w-full bg-white rounded-2xl shadow-xl p-6 sm:p-8 space-y-6">

        <!-- Title Section -->
        <div class="text-center">
            <h1 class="text-3xl sm:text-4xl font-extrabold text-gray-900 mb-2">सङ्गणकभाषा-अनुवादकम् (उन्नतम्)</h1>
            <p class="text-gray-500 text-sm sm:text-base">An advanced prototype with functions and classes.</p>
        </div>

        <!-- Code Editor -->
        <div>
            <label for="code-editor" class="block text-sm font-medium text-gray-700 mb-2">Code (कोड):</label>
            <textarea id="code-editor" class="w-full rounded-lg border border-gray-300 p-4 focus:outline-none focus:ring-2 focus:ring-indigo-500" placeholder="Type your Sanskrit code here...">
मुद्रयतु "कार्यम् प्रयोगम्"
कार्यम् वर्धयतु संख्या १, संख्या २
  मुद्रयतु "योगः: " + (संख्या १ + संख्या २)
समाप्ति
वर्धयतु १०, २०

मुद्रयतु "वर्गः प्रयोगम्"
वर्गः व्यक्तिः
  स्थिरम् नाम
  स्थिरम् वयः
  निर्मातम् नाम, वयः
    अहम्.नाम = नाम
    अहम्.वयः = वयः
  समाप्ति
  कार्यम् परिचयम्
    मुद्रयतु "अहम् " + अहम्.नाम + ", मम वयः " + अहम्.वयः + " अस्ति।"
  समाप्ति
समाप्ति

स्थिरम् पुरुषः = व्यक्तिः नवम् "राम", २५
पुरुषः.परिचयम्
            </textarea>
        </div>

        <!-- Buttons -->
        <div class="flex flex-col sm:flex-row space-y-4 sm:space-y-0 sm:space-x-4 justify-center">
            <button id="run-button" class="bg-indigo-600 hover:bg-indigo-700 text-white font-bold py-3 px-6 rounded-lg shadow-md transition-transform transform hover:scale-105 focus:outline-none focus:ring-2 focus:ring-indigo-500 focus:ring-opacity-50">
                <i class="fas fa-play mr-2"></i> निष्पादनम् (Run)
            </button>
            <button id="clear-button" class="bg-red-500 hover:bg-red-600 text-white font-bold py-3 px-6 rounded-lg shadow-md transition-transform transform hover:scale-105 focus:outline-none focus:ring-2 focus:ring-red-500 focus:ring-opacity-50">
                <i class="fas fa-trash-alt mr-2"></i> स्पष्टम् (Clear)
            </button>
        </div>

        <!-- Console Output -->
        <div>
            <label for="console-output" class="block text-sm font-medium text-gray-700 mb-2">Console (कन्सोल्):</label>
            <pre id="console-output" class="rounded-lg p-4 font-mono text-sm"></pre>
        </div>

    </div>

    <script>
        const codeEditor = document.getElementById('code-editor');
        const runButton = document.getElementById('run-button');
        const clearButton = document.getElementById('clear-button');
        const consoleOutput = document.getElementById('console-output');

        const variables = new Map();
        const functions = new Map();
        const classes = new Map();

        function print(message) {
            consoleOutput.textContent += message + '\\n';
            consoleOutput.scrollTop = consoleOutput.scrollHeight;
        }

        // --- Core Interpreter Logic ---
        function runCode(code) {
            const lines = code.split('\\n').map(line => line.trim()).filter(line => line.length > 0);
            let i = 0;
            while (i < lines.length) {
                const line = lines[i];
                const parts = line.split(/\\s+/);
                const command = parts[0];

                if (command === 'कार्यम्') {
                    const funcName = parts[1];
                    const params = parts.slice(2).join('').split(',').map(p => p.trim());
                    const body = [];
                    i++;
                    while (i < lines.length && lines[i] !== 'समाप्ति') {
                        body.push(lines[i]);
                        i++;
                    }
                    functions.set(funcName, { params, body });
                } else if (command === 'वर्गः') {
                    const className = parts[1];
                    const classDefinition = { properties: new Map(), methods: new Map(), constructor: null };
                    i++;
                    while (i < lines.length && lines[i] !== 'समाप्ति') {
                        const classLine = lines[i];
                        const classParts = classLine.split(/\\s+/);
                        if (classParts[0] === 'स्थिरम्') {
                            classDefinition.properties.set(classParts[1], null);
                        } else if (classParts[0] === 'निर्मातम्') {
                            const params = classParts.slice(1).join('').split(',').map(p => p.trim());
                            const body = [];
                            i++;
                            while (i < lines.length && lines[i] !== 'समाप्ति') {
                                body.push(lines[i]);
                                i++;
                            }
                            classDefinition.constructor = { params, body };
                            // No break here, need to parse methods after constructor
                        } else if (classParts[0] === 'कार्यम्') {
                            const methodName = classParts[1];
                            const params = classParts.slice(2).join('').split(',').map(p => p.trim());
                            const body = [];
                            i++;
                            while (i < lines.length && lines[i] !== 'समाप्ति') {
                                body.push(lines[i]);
                                i++;
                            }
                            classDefinition.methods.set(methodName, { params, body });
                        }
                        i++;
                    }
                    classes.set(className, classDefinition);
                } else {
                    executeLine(line);
                }
                i++;
            }
        }

        function executeLine(line, scope = variables) {
            const parts = line.split(/\\s+/);
            const command = parts[0];

            switch(command) {
                case 'मुद्रयतु':
                    const messageParts = line.split(/\\s+/).slice(1);
                    const messageEval = messageParts.join(' ').replace(/\\s?\\+\\s?/g, '+');
                    try {
                        const evaluatedMessage = eval(`(function() {
                            ${Array.from(scope.keys()).map(key => `const ${key} = scope.get("${key}");`).join('\\n')}
                            return ${messageEval};
                        })()`);
                        // Handle potential string concatenation with non-strings after eval
                        print(String(evaluatedMessage));
                    } catch (e) {
                        print(`ERROR: Syntax error or variable not found: ${e.message}`);
                    }
                    break;

                case 'स्थिरम्':
                case 'परिवर्तनीयम्':
                    const varName = parts[1];
                    const eqIndex = parts.indexOf('=');
                    if (eqIndex !== -1) {
                        const valueStr = parts.slice(eqIndex + 1).join(' ');
                        try {
                             // Handle potential string values correctly
                             const value = eval(valueStr);
                            scope.set(varName, value);
                        } catch (e) {
                             print(`ERROR: Invalid value for variable '${varName}': ${e.message}`);
                        }
                    } else {
                        print(`ERROR: Invalid variable assignment syntax for '${line}'`);
                    }
                    break;

                case 'यदि':
                    // Simplified: Just acknowledge the command
                    print(`  Acknowledging 'यदि' command.`);
                    break;

                case 'पुनरावर्तनम्':
                    // Simplified: Just acknowledge the command
                     print(`  Acknowledging 'पुनरावर्तनम्' command.`);
                    break;


                default:
                     // Handle function calls
                    if (functions.has(command)) {
                        const func = functions.get(command);
                        const argsStr = parts.slice(1).join('');
                        const args = argsStr.split(',').map(arg => {
                           try {
                                // Try to evaluate arguments (numbers, strings, variables)
                               return eval(arg);
                           } catch (e) {
                                print(`WARNING: Could not evaluate argument '${arg}'. Treating as string.`);
                               return arg; // Treat as string if eval fails
                           }
                        });

                        if (func.params.length === args.length) {
                             print(`Calling function '${command}' with arguments: ${args.join(', ')}`);
                             const funcScope = new Map(variables); // Inherit global variables
                             func.params.forEach((param, index) => funcScope.set(param, args[index]));
                             func.body.forEach(line => executeLine(line, funcScope)); // Execute function body
                        } else {
                             print(`ERROR: Incorrect number of arguments for function '${command}'. Expected ${func.params.length}, got ${args.length}.`);
                        }
                    } else if (line.includes('.')) { // Handle method calls
                         const [objName, methodCall] = line.split('.');
                         const instance = variables.get(objName);
                         if (instance && instance.__class__ && classes.has(instance.__class__)) {
                            const classDef = classes.get(instance.__class__);
                            const [methodName, ...args] = methodCall.split(/\\s+/); // Simpler split for method name and args
                            const methodNameClean = methodName.replace('()', ''); // Remove potential ()

                            if (classDef.methods.has(methodNameClean)) {
                                const method = classDef.methods.get(methodNameClean);
                                 // Basic method execution - in a real interpreter, this would be more complex
                                 print(`Executing method '${methodNameClean}' on object '${objName}'...`);
                                 // A real implementation would need to parse and execute the method body in the object's scope
                                 // For this simple demo, we just print a message.
                            } else {
                                 print(`ERROR: Method '${methodNameClean}' not found on object '${objName}' of class '${instance.__class__}'.`);
                            }
                         } else {
                            print(`ERROR: Object '${objName}' not found or is not a valid class instance.`);
                         }
                    } else if (line.trim() !== '') {
                        print(`ERROR: Unknown command or syntax: '${line}'`);
                    }
                    break;
            }
        }

        runButton.addEventListener('click', () => {
            consoleOutput.textContent = '';
            variables.clear();
            functions.clear();
            classes.clear();
            runCode(codeEditor.value);
            print('Program finished.');
        });

        clearButton.addEventListener('click', () => {
            codeEditor.value = '';
            consoleOutput.textContent = '';
        });
    </script>
</body>
</html>
"""

display(HTML(html_code))
```
